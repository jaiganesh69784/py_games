This program is a simple number guessing game where the user selects a range of numbers and can choose to have limited attempts to guess the randomly generated number within that range.

```python
import random

def guess_number(start, end, max_attempts=None):
    num = random.randint(start, end)
    count = 0
    if max_attempts is None:
        print("You have unlimited attempts.")
    else:
        print(f"You have {max_attempts} attempts.")

    while True:
        guess = input(f"Enter your guess number ({start}-{end}) or 'q' to quit: ")
        if guess.lower() == 'q':
            print("You have quit the game.")
            return
        try:
            guess = int(guess)
        except ValueError:
            print("Invalid input. Please enter a number.")
            continue
        count += 1
        if guess == num:
            if count == 1:
                print("Bang! You are amazing! Guessed it in the first attempt!")
            elif count <= 5:
                print("Wow! You got it within 5 attempts!")
            elif count <= 15:
                print("Good job! You guessed it within 15 attempts!")
            else:
                print("It took some time, but you eventually got it!")
            return
        elif guess > num:
            print(guess, "is High")
        else:
            print(guess, "is Low")
        if max_attempts is not None and count >= max_attempts:
            print("Out of moves! The correct number was", num)
            return

def main():  # Corrected the function name
    print("Welcome to the Number Guessing Game!")
    print("Choose the range of numbers:")
    print("1. 0-10")
    print("2. 0-100")
    print("3. 0-500")
    print("4. 0-1000")
    choice = input("Enter your choice (1-4): ")

    if choice == '1':
        start, end = 0, 10
    elif choice == '2':
        start, end = 0, 100
    elif choice == '3':
        start, end = 0, 500
    elif choice == '4':
        start, end = 0, 1000
    else:
        print("Invalid choice. Please choose a number between 1 and 4.")
        return

    attempts_choice = input("Do you want limited attempts? (yes/no): ")
    if attempts_choice.lower() == 'yes':
        max_attempts = int(input("Enter the number of attempts: "))
        guess_number(start, end, max_attempts)
    elif attempts_choice.lower() == 'no':
        guess_number(start, end)
    else:
        print("Invalid choice. Please enter 'yes' or 'no'.")

if __name__ == "__main__":
    main()
```
